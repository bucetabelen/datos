```python
## Análisis de valores nulos

# Se analizó la presencia de valores faltantes en el dataset utilizando la función `isnull()`.

# Se observa que varias columnas presentan una gran cantidad de valores nulos. En particular, algunas columnas poseen valores nulos en la totalidad de los registros, lo que indica que no contienen información útil para el análisis.

# Entre estas columnas se encuentran variables como:

# - `price`
# - `host_since`
# - `host_response_rate`
# - `host_acceptance_rate`
# - `instant_bookable`
# - `neighbourhood_group_cleansed`

# Estas variables no aportan información en el dataset analizado y podrían eliminarse en un proceso de limpieza de datos posterior.
```

```python
""" Los alojamientos con mas reviews tienen mas rating? """
sns.scatterplot(
    data=df,
    x="number_of_reviews",
    y="review_scores_rating"
)

plt.title("Relación entre cantidad de reviews y rating")
plt.xlabel("Número de reviews")
plt.ylabel("Rating")
plt.show()
```

Se analizó la relación entre la cantidad de reviews y el rating promedio de los alojamientos.

El gráfico de dispersión permite observar si existe alguna relación entre la popularidad de un alojamiento (medida por el número de reviews) y la valoración que reciben por parte de los huéspedes.

```python
review_scores = df[[
    "review_scores_rating",
    "review_scores_accuracy",
    "review_scores_cleanliness",
    "review_scores_checkin",
    "review_scores_communication",
    "review_scores_location",
    "review_scores_value"
]]
correlation_matrix = review_scores.corr()
```

```python
plt.figure(figsize=(8,6))

sns.heatmap(
    correlation_matrix,
    annot=True,
    cmap="coolwarm",
    fmt=".2f"
)

plt.title("Correlación entre distintos scores de reviews")
plt.show()
```

Se analizó la correlación entre los distintos puntajes de evaluación de los alojamientos.

El heatmap permite visualizar qué aspectos del alojamiento (limpieza, ubicación, comunicación, etc.) presentan mayor relación con el rating general otorgado por los huéspedes.

```python
corr = df.select_dtypes(include=["int64","float64"]).corr()

plt.figure(figsize=(10,8))

sns.heatmap(
    corr,
    cmap="coolwarm",
    annot=False
)

plt.title("Matriz de correlación entre variables numéricas")
plt.show()
```
