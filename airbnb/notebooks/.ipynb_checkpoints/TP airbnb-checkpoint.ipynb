```python
import matplotlib.pyplot as plt

df1["room_type"].value_counts().plot(kind="bar")

plt.title("Distribución de tipos de alojamiento")
plt.xlabel("Tipo de alojamiento")
plt.ylabel("Cantidad de propiedades")

plt.xticks(rotation=45)

plt.show()
```

```python
df1["neighbourhood_cleansed"].value_counts().head(10).plot(kind="bar")

plt.title("Top 10 barrios con mayor cantidad de alojamientos")
plt.xlabel("Barrio")
plt.ylabel("Cantidad de propiedades")

plt.xticks(rotation=45)

plt.show()
```

```python
import matplotlib.pyplot as plt

counts = df1["host_is_superhost"].value_counts()

color_map = {
    "t": "green",
    "f": "red"
}

colors = [color_map[v] for v in counts.index]

counts.plot(
    kind="pie",
    autopct="%1.1f%%",
    startangle=90,
    colors=colors
)

plt.title("Proporción de Superhosts")
plt.ylabel("")
plt.show()
```

```python
counts = df1["host_identity_verified"].value_counts()

colors = [color_map[v] for v in counts.index]

counts.plot(
    kind="pie",
    autopct="%1.1f%%",
    startangle=90,
    colors=colors
)

plt.title("Hosts con identidad verificada")
plt.ylabel("")
plt.show()
```

```python
porcentaje = df1["room_type"].value_counts(normalize=True) * 100

ax = porcentaje.plot(kind="bar")

for i, v in enumerate(porcentaje):
    ax.text(i, v + 0.5, f"{v:.1f}%", ha="center")

plt.title("Distribución de tipos de alojamiento (%)")
plt.ylabel("Porcentaje")
plt.xlabel("Tipo de alojamiento")

plt.xticks(rotation=45)

plt.show()
```
